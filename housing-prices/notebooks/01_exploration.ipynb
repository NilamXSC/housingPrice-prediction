```python
# Cell 1: Imports & load data
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', 120)
pd.set_option('display.max_rows', 200)
sns.set(style="whitegrid")

DATA_PATH = r"F:\Data Science\housing-prices\data\train.csv"
assert os.path.exists(DATA_PATH), f"Data file not found at {DATA_PATH}"

df = pd.read_csv(DATA_PATH)
print("Loaded dataset:", DATA_PATH)
print("Shape:", df.shape)
```

```python
# Cell 2: Quick glance
# Show top rows
display(df.head())

# Data types and non-null counts
print("\nInfo:")
df.info()

# Summary statistics for numeric columns (transpose for readability)
print("\nNumeric summary (first 20 rows):")
display(df.describe().T.head(20))
```

```python
# Cell 3: Missing values
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_df = missing_df[missing_df['missing_count'] > 0]
display(missing_df.head(50))
```

```python
# Cell 4: Categorical columns overview
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
print(f"Categorical columns ({len(cat_cols)}): {cat_cols[:15]}{'...' if len(cat_cols)>15 else ''}")

# Show unique counts for the top categorical columns (first 20)
cat_uniques = df[cat_cols].nunique().sort_values(ascending=False)
display(cat_uniques.head(50))
```

```python
# Cell 5: Target distribution
target = 'SalePrice'
if target not in df.columns:
    raise ValueError("SalePrice column not found in dataframe.")

plt.figure(figsize=(10,4))
sns.histplot(df[target], kde=True)
plt.title('SalePrice distribution (raw)')
plt.xlabel('SalePrice')
plt.show()

plt.figure(figsize=(10,4))
sns.boxplot(x=df[target])
plt.title('SalePrice boxplot (raw)')
plt.show()

# Log transform
plt.figure(figsize=(10,4))
sns.histplot(np.log1p(df[target]), kde=True)
plt.title('Log(1 + SalePrice) distribution')
plt.show()
```

```python
# Cell 6: Numeric correlations with target
num_corr = df.corr(numeric_only=True)
corr_with_target = num_corr[target].sort_values(ascending=False)
display(corr_with_target.head(20))
display(corr_with_target.tail(20))
```

```python
# Cell 7: Scatter plots for selected top predictors
top_feats = corr_with_target.index.drop(target).tolist()[:6]  # top 6 numeric features correlated with SalePrice
print("Top numeric correlated features:", top_feats)

for f in top_feats:
    plt.figure(figsize=(6,3))
    sns.scatterplot(x=df[f], y=df[target])
    plt.title(f'SalePrice vs {f}')
    plt.xlabel(f)
    plt.ylabel('SalePrice')
    plt.tight_layout()
    plt.show()
```

```python
# Cell 8: SalePrice by Neighborhood (boxplot)
if 'Neighborhood' in df.columns:
    order = df.groupby('Neighborhood')['SalePrice'].median().sort_values().index
    plt.figure(figsize=(14,6))
    sns.boxplot(data=df, x='Neighborhood', y='SalePrice', order=order)
    plt.xticks(rotation=90)
    plt.title('SalePrice by Neighborhood (ordered by median)')
    plt.show()
else:
    print("No 'Neighborhood' column found.")
```

```python
# Cell 9: Outlier detection for a few important numeric cols
check_cols = ['GrLivArea', 'TotalBsmtSF', 'GarageArea', 'LotArea']
existing_cols = [c for c in check_cols if c in df.columns]

for c in existing_cols:
    q1 = df[c].quantile(0.01)
    q99 = df[c].quantile(0.99)
    very_small = df[df[c] < q1].shape[0]
    very_large = df[df[c] > q99].shape[0]
    print(f"{c}: 1st pctile={q1:.1f}, 99th pctile={q99:.1f}, <1pct: {very_small}, >99pct: {very_large}")

# Show rows with extremely large living area (example)
if 'GrLivArea' in df.columns:
    large = df[df['GrLivArea'] > df['GrLivArea'].quantile(0.995)]
    if not large.empty:
        print("\nRows with very large GrLivArea (top 0.5%):")
        display(large[['Id','GrLivArea','SalePrice']].sort_values('GrLivArea', ascending=False).head(10))
```

```python
# Cell 10: heatmap for top numeric features
# pick top 12 numeric features most correlated to SalePrice
top_numeric = corr_with_target.drop(labels=[target]).abs().sort_values(ascending=False).head(12).index.tolist()

plt.figure(figsize=(10,8))
sns.heatmap(df[top_numeric + [target]].corr(numeric_only=True), annot=True, fmt=".2f", cmap="RdYlBu")
plt.title("Correlation matrix (top numeric features + SalePrice)")
plt.show()
```

```python
# Cell 11: preview neighborhood aggregate (useful later for feature engineering)
if 'Neighborhood' in df.columns:
    neigh_price = df.groupby('Neighborhood')['SalePrice'].median().rename('Neighborhood_MedianPrice')
    neigh_df = df[['Neighborhood']].merge(neigh_price, left_on='Neighborhood', right_index=True)
    display(neigh_df.head().assign(Neighborhood_MedianPrice=lambda x: x['Neighborhood_MedianPrice'].astype(int)))
else:
    print("No 'Neighborhood' column found.")
```

```python
# Cell 12: save EDA snapshot: missing, top correlations, and head
out_dir = r"F:\Data Science\housing-prices\reports"
os.makedirs(out_dir, exist_ok=True)

missing_df.to_csv(os.path.join(out_dir, "missing_values_summary.csv"))
corr_with_target.to_csv(os.path.join(out_dir, "corr_with_saleprice.csv"))
df.head(100).to_csv(os.path.join(out_dir, "data_head_100.csv"), index=False)

print("Saved EDA snapshots to", out_dir)
print("\nNEXT STEPS SUGGESTION:\n- Decide if you want to model SalePrice or log(SalePrice) (log recommended).\n- Handle missing values: some NaNs mean 'None' (e.g., PoolQC, Alley), others need imputation.\n- Engineer features: house_age, total_sqft, total_bathrooms, neighborhood median price, etc.\n- Build preprocessing pipeline & baseline model (we will do that next).")
```
